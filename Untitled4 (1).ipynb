```python
def predict_using_sklearn():
    df = pd.read_csv("test_scores.csv")
    
    r = LinearRegression()
    
    r.fit(df[['math']], df['cs'])
    
    return r.coef_[0], r.intercept_

def gradient_descent(x, y):
    m_curr = 0  
    b_curr = 0  
    iterations = 1000000  
    n = len(x)  
    learning_rate = 0.0002
    cost_previous = float('inf') 
    tolerance = 1e-6 

    for i in range(iterations):
        y_predicted = m_curr * x + b_curr 
        cost = (1/n) * sum([value**2 for value in (y - y_predicted)])  

      
        md = -(2/n) * sum(x * (y - y_predicted)) 
        bd = -(2/n) * sum(y - y_predicted)

       
        m_curr -= learning_rate * md
        b_curr -= learning_rate * bd

   
        if abs(cost - cost_previous) < tolerance:
            break
        
        cost_previous = cost  
       
        if i % 10000 == 0:
            print(f"Iteration {i}: m = {m_curr}, b = {b_curr}, cost = {cost}")

    return m_curr, b_curr

if __name__ == "__main__":
    df = pd.read_csv("test_scores.csv")
    x = np.array(df['math'])
    y = np.array(df['cs'])  

    m, b = gradient_descent(x, y)
    print(f"Using gradient descent: Coef = {m}, Intercept = {b}")

    m_sklearn, b_sklearn = predict_using_sklearn()
    print(f"Using sklearn: Coef = {m_sklearn}, Intercept = {b_sklearn}")
```
